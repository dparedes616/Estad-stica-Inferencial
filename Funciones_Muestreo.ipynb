### Funcion de muestreo sistematico

```python
def systematic_sampling(df_raw,step):
    indexes =np.arange(0, len(df_raw),step=step)
    systematic_sample = df_raw.iloc[indexes]
    return systematic_sample

systematic_sample = systematic_sampling(df_raw,3)
systematic_sample
```

### Muestre de estratificación

```python
df_raw["estratificado"]=df_raw["delegacion"]+ "," + df_raw["tipo"]
round(((df_raw["estratificado"].value_counts()/len(df_raw))*100),2)
```

```python
def data_estratificada(df_raw,nombres_columnas_estrat,valores_estrat,prop_estrat,random_state=None):
    df_estrat = pd.DataFrame(columns= df_raw.columns)

    pos=-1
    for i in range(len(valores_estrat)):
        pos =+1
        if pos == len(valores_estrat)-1:
            ratio_len = len(df_raw)-len(df_estrat)
        else:
            ratio_len = int(len(df_raw)*prop_estrat[i])
        
        df_filtrado= df_raw[df_raw[nombres_columnas_estrat]==valores_estrat[i]]
        df_temp = df_filtrado.sample(replace=True, n=ratio_len, random_state=random_state)

        df_estrat =pd.concat([df_estrat, df_temp])
    return df_estrat
```

```python
valores_estrat = ['Cuautémoc,Hotel', 'Cuautémoc,Museo', 'Venustiano Carranza,Hotel', 'Cuauhtémoc,Mercado','Venustiano Carranza,Mercado']
prop_estrat = [0.5, 0.2, 0.1, 0.1,0.1]
df_estrat = data_estratificada(df_raw, 'estratificado', valores_estrat, prop_estrat, random_state=42)
df_estrat
```
